```python
from nltk.book import *
from nltk import FreqDist, bigrams
from collections import defaultdict

import matplotlib.pyplot as plt

# 1) Concordance examples (use different texts/words)
print("Concordance for 'whale' in text1:")
text1.concordance("whale")

print("\nConcordance for 'Elinor' in text2:")
text2.concordance("Elinor")

# 2) similar() and common_contexts()
print("\nWords similar to 'whale' in text1:")
text1.similar("whale")

print("\nWords similar to 'Elinor' in text2:")
text2.similar("Elinor")

print("\nCommon contexts for 'Elinor' and 'Marianne' in text2:")
text2.common_contexts(["Elinor", "Marianne"])

# 3) Dispersion plot (choose a text and words of interest)
plt.figure(figsize=(10, 3))
text4.dispersion_plot(["freedom", "citizens", "war", "peace"])  # replace words as desired
plt.show()

# 4) Lexical diversity (a short description and value)
def lexical_diversity(text):
    return len(text) / len(set(text))

print(f"\nLexical diversity of text2: {lexical_diversity(text2):.4f}")
# Lexical diversity is the ratio of total tokens to unique tokens (type-token ratio).

# 5) Frequency distribution for text2 and cumulative frequency plot for top 75 words
fd = FreqDist(text2)
print("\nTop 10 most common words in text2:", fd.most_common(10))

plt.figure(figsize=(12, 6))
fd.plot(75, cumulative=True)
plt.title("Cumulative frequency of top 75 words in text2")
plt.show()

# 6) Bigrams and collocations (what a bigram is and examples)
# A bigram is a pair of consecutive words. Collocations are frequent/meaningful word pairings.
bgs = list(bigrams(text2))
print("\nFirst 20 bigrams from text2:", bgs[:20])

print("\nCollocations found in text2:")
text2.collocations()

# 7) Short worked example related to exercise 2-12 (word-length frequency)
length_counts = defaultdict(int)
for w in text2:
    length_counts[len(w)] += 1
top_lengths = sorted(length_counts.items(), key=lambda x: x[1], reverse=True)[:10]
print("\nMost common word lengths in text2 (top 10):", top_lengths)
```

## Exercise 4

```python
# Exercise 4

from nltk.corpus import state_union
from collections import Counter
import matplotlib.pyplot as plt

# load fileids
fileids = state_union.fileids()  # ordered by date
counts = []
years = []
for fid in fileids:
    text = state_union.raw(fid).lower()
    c = Counter(nltk.word_tokenize(text))
    counts.append((c['men'], c['women'], c['people']))
    # fileid like '1945-Truman.txt' -> extract year
    year = int(fid.split('-')[0])
    years.append(year)

# Convert to lists
men = [t[0] for t in counts]
women = [t[1] for t in counts]
people = [t[2] for t in counts]

# # quick print for inspection (year, men,women,people)
# for y,m,w,p in zip(years, men, women, people)[-10:]:
#     print(y, m, w, p)

# (optional) plot
plt.plot(years, men, label='men')
plt.plot(years, women, label='women')
plt.plot(years, people, label='people')
plt.xlabel('Year')
plt.ylabel('Raw counts')
plt.legend()
plt.show()
```

### Interpretation / answer
This code produces time series for the raw token counts of men, women, and people in successive State of the Union addresses. Typical observations: usage of men is high in earlier decades, women is near-zero until later decades and then increases, while people rises in modern addresses as a gender-neutral collective term. (Plotting the series makes these trends clear.)

## Exercise 8

```python
# Exercise 8

from nltk.corpus import names
from nltk import ConditionalFreqDist

male_names = names.words('male.txt')
female_names = names.words('female.txt')

def initial_letter(name):
    return name[0].lower()

cfd = ConditionalFreqDist(
    (('male' if name in male_names else 'female'), initial_letter(name))
    for name in (male_names + female_names)
)

# example: letters with top counts for each sex
for gender in ['male', 'female']:
    print(gender, sorted(cfd[gender].items(), key=lambda t: -t[1])[:10])
```

### Interpretation / answer
This gives the distribution over initial letters for male vs female names. You can visualize with a bar chart or compute relative frequencies to see which initials are more characteristic of male vs female names (e.g., j, m are frequent for both, but relative ranks differ).

## Exercise 11

```python
# Exercise 11 

from nltk.corpus import brown
modals = ['can','could','may','might','must','will','would','shall','should']
genres = brown.categories()

# build table: genre -> freq of modal per 1000 tokens
import pandas as pd
rows = []
for g in genres:
    words = [w.lower() for w in brown.words(categories=g)]
    N = len(words)
    freqs = {m: words.count(m) / N * 1000 for m in modals}
    freqs['genre'] = g
    rows.append(freqs)

df = pd.DataFrame(rows).set_index('genre')
df.sort_values('will', ascending=False).head()  # example
```

### Interpretation / answer
Look for genres where deontic modals like must/should are concentrated (e.g., editorial, religious texts), while conversational genres have more can/will/like. Closed-class differences reflect genre conventions: news/editorials use certain modals more for obligation/necessity; fiction uses more will/would for future/conditional constructions.

## Exercise 16

```python
# Exercise 16

import pandas as pd
from nltk.corpus import brown

def lexical_diversity(cat):
    tokens = brown.words(categories=cat)
    return len(tokens) / len(set(w.lower() for w in tokens))

rows = []
for cat in brown.categories():
    rows.append({'genre': cat, 'tokens': len(brown.words(categories=cat)),
                 'types': len(set(w.lower() for w in brown.words(categories=cat))),
                 'tokens_per_type': lexical_diversity(cat)})
df = pd.DataFrame(rows).set_index('genre')
df.sort_values('tokens_per_type', ascending=False).head()
```

### Interpretation / answer
The genre with the lowest diversity has the largest tokens-per-type value (i.e., more repetition, fewer distinct words). Commonly youll find news/editorial or technical genres may have lower lexical diversity (higher tokens/type) than fiction or academic, but the exact ranking depends on the Brown split. ## Exercise 23

```python
# Exercise 23

import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
from nltk.corpus import gutenberg
from nltk.tokenize import word_tokenize
import random
import string

def zipf_counts(text_tokens):
    c = Counter([w.lower() for w in text_tokens if w.isalpha()])
    freqs = np.array(sorted(c.values(), reverse=True))
    ranks = np.arange(1, len(freqs)+1)
    return ranks, freqs

# choose a large text (e.g., 'melville-moby_dick.txt') or combine multiple Gutenberg files
text_tokens = word_tokenize(gutenberg.raw('melville-moby_dick.txt'))
ranks, freqs = zipf_counts(text_tokens)

plt.figure()
plt.loglog(ranks, freqs)
plt.xlabel('Rank (log)')
plt.ylabel('Frequency (log)')
plt.title('Zipf plot  Moby Dick (log-log)')
plt.show()

# Random-character text
chars = ''.join(random.choice("abcdefg ") for _ in range(5_000_00))  # long
rand_tokens = word_tokenize(chars)
ranks_r, freqs_r = zipf_counts(rand_tokens)

plt.figure()
plt.loglog(ranks_r, freqs_r)
plt.xlabel('Rank (log)')
plt.ylabel('Frequency (log)')
plt.title('Zipf plot  Random characters (log-log)')
plt.show()
```

### Observations / answer
* For natural text (e.g., Moby Dick), the rank vs frequency plot on a loglog scale is approximately a straight line over a wide range  consistent with Zipfs law (frequency  1 / rank^s with s  1). The extremes deviate: the very top ranks (the very most frequent function words) and the very low-frequency tail (hapax legomena) depart from the pure power-law.
* For random-character-generated words, the distribution looks markedly different  there is much less of a clean power-law line because the process does not create the strong long-tail structure present in human language.
* Conclusion: Zipfs law is not a trivial artifact of any random sequence  it reflects structure in language (and in many systems) and is typically associated with generative processes that produce heavy tails (e.g., preferential attachment, multiplicative processes, sampling from heavy-tailed distributions).

### Which property of BarabsiAlbert networks connects to Zipfs law?
The BarabsiAlbert (BA) model produces scale-free networks whose node degree distribution follows a power law: $P(k) \propto k^{-\gamma}$ (commonly $\gamma \approx 3$ in the simplest BA model). Zipfs law (rank vs frequency  power-law) is another manifestation of a power-law / heavy-tail distribution. Thus the connection is that Zipfs law for ranked frequencies is algebraically equivalent to a power-law distribution (so when node degrees follow a power law in BA networks, their ranked degrees also satisfy a Zipf-like relationship). In short: the BA models power-law degree distribution is the network analogue of Zipfs law for rank-frequency data.  

Important internet instances of power-laws (brief paragraph)
Researchers have observed power-law distributions throughout Internet phenomena: the degrees (number of links) of nodes in web graphs and router-level topologies, popularity/rankings of websites or pages (few pages get huge traffic, many get little), number of email contacts per user, file sizes and request frequencies, and cascade sizes in viral spread. These heavy-tailed distributions have practical implications: they affect the robustness of networks (hubs make some attacks very damaging), cache and load balancing strategies (a small fraction of items accounts for most accesses), and the epidemiology of viruses and information cascades.
